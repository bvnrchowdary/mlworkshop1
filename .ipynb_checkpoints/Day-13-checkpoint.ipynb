```python
# import packages
import pandas as pd
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
print(iris)
```

```python
input_data=pd.DataFrame(iris.data,columns=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'])
input_data.head()
```

```python
output_data=pd.DataFrame(iris.target,columns=['target'])
output_data.head()
```

```python
input_data.shape
```

```python
output_data.shape
```

```python
input_data.info()
```

```python
output_data.info()
```

```python
input_data.isnull().sum()
```

```python
output_data.isnull().sum()
```

```python
#spliting the data testing and traing
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(input_data,output_data,test_size=30,random_state=3)
```

```python
# import required model
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier()
```

```python
dtc.fit(x_train,y_train)
```

```python
#predict the values
```

```python
pred=dtc.predict(x_test)
pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_test,pred)
```

```python
confusion_matrix(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```
