### Linear Regression with One feature
- Linear model is sum weighted predicted data to target values
- y=mx+c
    - y = target value/OUTPUT
    - m = slope
    - x = Input value
    - c = Coefficient/Intercept
        - m = (x-xmean)*(y-ymean)/(x-xmean)^2
        - c = ymean-(m*xmean)
    

## ML steps :
 - 1.Get or Load Data
 - 2.Preprocessing the data
 - 3.Define input and output
 - 4.Apply the model or algorithm
 - 5.pass the data to training and testing
 - 6.calculate the score or accuracy score
 
 

```python
# Get the libraries
import pandas as pd
import numpy as np
```

```python
# 1.get the data
df=pd.read_csv("Salary_Data.csv")
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
# 2.preprocessing
df.isnull().sum()
```

```python
# 3.define input and output
x=df[['YearsExperience']]
y=df['Salary']
```

```python
# 4.apply the model or algorithm
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(x,y)
```

```python
# 5.test the data
model.predict([[7.1],[10]])
```

```python
# 6. Identify the score
model.score(x,y)*100
```

```python
# visualize the relationship of input and output
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['YearsExperience'],df['Salary'],label='actual data')
plt.show()
```

```python
y_pred = model.predict(x)
```

```python
plt.scatter(df['YearsExperience'],df['Salary'],label='actual data')
plt.plot(x,y_pred,color='r',label="predicted values")
plt.title("Years of Experience vs Salary")
plt.xlabel("years of experiance")
plt.ylabel("salary")
plt.legend()
plt.show()
```
